```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
```

### Chrome WebDriver 실행

```python
# Chrome WebDriver 경로 설정
chrome_driver_path = "./chromedriver.exe" # 프로젝트와 같은 경로
service = Service(chrome_driver_path)

# WebDriver 실행
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # 백그라운드에서 실행
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(service=service, options=options)
```

### 멜론 차트 top 100 가져오기

```python
# 멜론 차트 URL
m_url = "https://www.melon.com/chart/index.htm"

# 멜론 웹사이트 열기
driver.get(m_url)
time.sleep(3)

# 데이터를 저장할 리스트 초기화
m_titles, m_artists = [], []

# 멜론 1~100 데이터 크롤링
for i in range(1, 101):
    try:
        title = driver.find_element(By.XPATH, f'//tr[{i}]/td[6]/div/div/div[1]/span/a').text
        artist = driver.find_element(By.XPATH, f'//tr[{i}]/td[6]/div/div/div[2]/a').text
        m_titles.append(title)
        m_artists.append(artist)
    except Exception as e:
        print(f"{i}위 데이터를 가져오는 중 오류 발생: {e}")

print(m_titles)
print(m_artists)
```

### 벅스 차트 top 100 가져오기

```python
# 벅스 차트 URL
b_url = "https://music.bugs.co.kr/chart"

# 벅스 웹사이트 열기
driver.get(b_url)
time.sleep(3)

# 데이터를 저장할 리스트 초기화
b_titles, b_artists = [], []

# 벅스 1~100 데이터 크롤링
for i in range(1, 101):
    try:
        title = driver.find_element(By.XPATH, f'//tr[{i}]/th/p/a').text
        artist = driver.find_element(By.XPATH, f'//tr[{i}]/td[5]/p/a').text
        b_titles.append(title)
        b_artists.append(artist)
    except Exception as e:
        print(f"{i}위 데이터를 가져오는 중 오류 발생: {e}")

print(b_titles)
print(b_artists)
```

### 지니 차트 top 100 가져오기
- 지니 차트의 경우 1~50위 단위로 페이지가 나뉘어져 있음
- for 문으로 페이지를 2번 열어서 가져오기

```python
# 지니 차트 URL
g_url = "https://www.genie.co.kr/chart/top200?ditc=D&ymd=20250123&hh=10&rtm=Y&pg="

# 데이터를 저장할 리스트 초기화
g_titles, g_artists = [], []

# 지니 웹사이트 열기
for i in range(1, 3):
    driver.get(g_url+str(i))
    time.sleep(3)

    # 지니 1~100 데이터 크롤링
    for j in range(1, 51):
        try:
            title = driver.find_element(By.XPATH, f'//tr[{j}]/td[5]/a').text
            artist = driver.find_element(By.XPATH, f'//tr[{j}]/td[5]/a[2]').text
            g_titles.append(title)
            g_artists.append(artist)
        except Exception as e:
            print(f"{(i-1)*50+j}위 데이터를 가져오는 중 오류 발생: {e}")

print(g_titles)
print(g_artists)
```

```python
# WebDriver 종료
driver.quit()
```

### 데이터 프레임 생성하기

```python
# 데이터 프레임 생성
service = ['Melon', 'Bugs', 'genie']
titles = [m_titles, b_titles, g_titles]
artists = [m_artists, b_artists, g_artists]
data = []

for i in range(3):
    data.append({'service':[service[i]]*100, 'ranking':range(1, 101), 'title':titles[i], 'artist': artists[i]})

melon = pd.DataFrame(data[0])
bugs = pd.DataFrame(data[1])
genie = pd.DataFrame(data[2])
```
